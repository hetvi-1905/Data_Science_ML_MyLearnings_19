```python
# groupby()
df = pd.DataFrame({'Animal': ['Falcon', 'Falcon',
                              'Parrot', 'Parrot'],
                   'Max Speed': [380., 370., 24., 26.]})
df
```

```python
df.groupby(['Animal']).mean()
```

```python
#We can groupby different levels of a hierarchical index using the level parameter:

arrays = [['Falcon', 'Falcon', 'Parrot', 'Parrot'],
          ['Captive', 'Wild', 'Captive', 'Wild']]
index = pd.MultiIndex.from_arrays(arrays, names=('Animal', 'Type'))
df = pd.DataFrame({'Max Speed': [390., 350., 30., 20.]},
                  index=index)
print(df)

print(df.groupby(level=0).mean())

print(df.groupby(level=1).mean())

print(df.groupby(level="Type").mean())

print(df.groupby(level="Animal").mean())

```

```python
# merge
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
print(df1)
print(df2)
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey')   # like cross product
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey',
          suffixes=('_left', '_right'))
```

```python
df1 = pd.DataFrame({'a': ['foo', 'bar'], 'b': [1, 2]})
df2 = pd.DataFrame({'a': ['foo', 'baz'], 'c': [3, 4]})
print(df1)
print(df2)
```

```python
df1.merge(df2, how='inner', on='a')                  #like inner join
```

```python
df1.merge(df2, how='left', on='a')                  # left outer join
```

```python
df1 = pd.DataFrame({'left': ['foo', 'bar']})
df2 = pd.DataFrame({'right': [7, 8]})
print(df1)
print(df2)
```

```python
df1.merge(df2, how='cross')   # cross product
```

#  Drop NA

```python
# dropna()

df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
df
```

```python
#Drop the rows where at least one element is missing.

df.dropna()
```

```python
#Drop the columns where at least one element is missing.

df.dropna(axis='columns')  # or
df.dropna(axis=1)
```

```python
#Drop the rows where all elements are missing.

df.dropna(how='all')
```

```python
#Keep only the rows with at least 2 non-NA values.

df.dropna(thresh=2)
```

```python
#Keep only the cols(axis=1) with at least 2 non-NA values.

df.dropna(thresh=2,axis=1)
```

```python
#Define in which columns to look for missing values.

df.dropna(subset=['name', 'toy'])
```

```python
#Keep the DataFrame with valid entries in the same variable.

df.dropna(inplace=True)
df
```

# Fill na

```python
# fillna()
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan, 4]],
                  columns=list("ABCD"))
df
```

```python
#Replace all NaN elements with 0s.

df.fillna(0)
```

```python
#We can also propagate "non-null" values forward or backward.

df.fillna(method="ffill")
```

```python
#Replace all NaN elements in column A, B, C, and D, with 0, 1, 2, and 3 respectively.

values = {"A": 0, "B": 1, "C": 2, "D": 3}
df.fillna(value=values)
```

```python
#Only replace the first NaN element.

df.fillna(value=values, limit=1)
```

```python
#When filling using a DataFrame, replacement happens along the same column names and same indices

df2 = pd.DataFrame(np.zeros((4, 4)), columns=list("ABCE"))
print(df)
print(df2)
df.fillna(df2)
```
